```python
def points_inside_polygon(points):
    point_x_gpu = cudf.Series(points["latitude"])
    point_y_gpu = cudf.Series(points["longitude"])
    result = cuspatial.point_in_polygon_bitmap(
    point_x_gpu, # points_x
    point_y_gpu, # points_y
    cudf.Series([1]), # one polygon
    polygon_last_point, # on what position (index+1, not just index!) each polygon ends
    polygon_x_gpu, # polygons_x
    polygon_y_gpu # polygons_y
)
    return result
```
